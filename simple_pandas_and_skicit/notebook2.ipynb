```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('tvmarketing.csv')
X=data[['TV']]
y=data['Sales']
```

```python
scaler=StandardScaler()
```

```python
X_norm=scaler.fit_transform(X)
linr_reg=SGDRegressor(max_iter=2000)
linr_reg.fit(X_norm,y)
```

```python
b_norm = linr_reg.intercept_
w_norm = linr_reg.coef_
print(b_norm,"\n",w_norm)
```

```python
plt.scatter(X, y)
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('TV Advertising vs Sales')
plt.show()
```

```python
k=linr_reg.predict(X_norm)
```

```python
plt.scatter(X,k)
plt.scatter(X,y)
plt.show()
```
