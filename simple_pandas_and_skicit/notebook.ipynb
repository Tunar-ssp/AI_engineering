```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("rideshare_2022.csv", parse_dates=['Trip Start Timestamp', 'Trip End Timestamp'])

df.head()
```

```python
columns_of_interest = ['Trip Start Timestamp', 'Trip Seconds',
       'Trip Miles', 'Fare', 'Tip', 'Additional Charges', 'Trip Total', 'Shared Trip Authorized',
       'Trips Pooled', 'Pickup Centroid Latitude', 'Pickup Centroid Longitude', 'Dropoff Centroid Latitude',
       'Dropoff Centroid Longitude']
df=df[columns_of_interest]
```

```python
rename_columns={i: "_".join(  (i.lower()).split(" ")  ) for i in df.columns}

df.rename(columns=rename_columns,inplace=True)
df['date'] = pd.to_datetime(df['trip_start_timestamp'].dt.date)
```

```python
df.info()
```

```python
WEEKDAYS = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['weekday'] = df["date"].dt.day_name()
```

```python
daily_rider=df['weekday'].value_counts()

daily_rider=daily_rider.reindex(WEEKDAYS)

daily_rider
```

```python
df_tippers=df[df['tip']>0]

daily_tipper_count=df_tippers['weekday'].value_counts()

daily_tipper_count= daily_tipper_count.reindex(WEEKDAYS)

print(daily_tipper_count)

```

```python
df_daily_agregetion=pd.concat(
    [daily_rider,daily_tipper_count],
    axis=1,
    keys=['rider_count','tipper_count']
)
df_daily_agregetion["tipper_percentage"]=df_daily_agregetion['tipper_count']/df_daily_agregetion['rider_count']*100


df_daily_agregetion
```

```python
df_daily_agregetion.to_csv('daily_agregetion_data.csv')
```
